**Efetividade de Duas Estratégias de Ensino**

Imagine que uma escola esteja avaliando a eficácia de duas estratégias de ensino de matemática para alunos do ensino médio. Eles querem determinar se há uma diferença significativa no desempenho médio dos alunos entre as duas estratégias.

# **Hipóteses:**

* Hipótese nula (H0): A média das notas dos alunos na estratégia A é igual à média das notas dos alunos na estratégia B.
* Hipótese alternativa (H1): A média das notas na Estratégia B é maior do que a média das notas na Estratégia A.

# **Dados:**

* Amostra da Estratégia A: Notas de 50 alunos que receberam a Estratégia A.
* Amostra da Estratégia B: Notas de 50 alunos que receberam a Estratégia B.

Usaremos um teste Z para comparar as médias das notas entre as duas amostras.

Se o p-valor do teste Z for menor que um nível de significância pré-determinado (por exemplo, α = 0.05), rejeitamos a hipótese nula e concluímos que há uma diferença significativa nas médias das notas entre as duas estratégias de ensino.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

Os dados são criados a seguir:

```python
# Definindo médias e desvios padrão para as notas nas duas estratégias
media_estrategia_A = 70
desvio_padrao_estrategia_A = 10

media_estrategia_B = 75
desvio_padrao_estrategia_B = 12

# Gerando as amostras de notas para cada estratégia de ensino da nossa base
np.random.seed(0)  # Para garantir a reprodutibilidade dos resultados
amostra_estrategia_A = np.random.normal(loc=media_estrategia_A, scale=desvio_padrao_estrategia_A, size=50)
amostra_estrategia_B = np.random.normal(loc=media_estrategia_B, scale=desvio_padrao_estrategia_B, size=50)

print("Notas da Estratégia A:", amostra_estrategia_A[:5])
print("Notas da Estratégia B:", amostra_estrategia_B[:5])
```

# 1) De acordo com as informações analisadas o nosso teste é unilateral á direita, esquerda ou bicaudal? Justifique.




Temos que a hipótese nula $H_0$ afirma que a média das notas na Estratégia B é menor ou igual à média das notas na Estratégia A. Como queremos verificar que a média da Estratégia B é maior, utilizamos o teste unilateral à *direita*.


# 2) Calcule as médias para as duas amostragens e as variâncias. Quais *insights* você pode retirar comparando os dados?

```python
media_amostral_estrategia_A = amostra_estrategia_A.mean()
media_amostral_estrategia_B = amostra_estrategia_B.mean()

variancia_amostral_estrategia_A = amostra_estrategia_A.var(ddof=1)
variancia_amostral_estrategia_B = amostra_estrategia_B.var(ddof=1)

print("Média amostral de A: ", media_amostral_estrategia_A.round(3))
print("Média amostral de B: ", media_amostral_estrategia_B.round(3))
print("Diferença absoluta entre as médias amostrais: ", abs(media_amostral_estrategia_A - media_amostral_estrategia_B).round(3))
print()
print("Variância de A amostral: ", variancia_amostral_estrategia_A.round(3))
print("Variância de B amostral: ", variancia_amostral_estrategia_B.round(3))
print("Diferença absoluta entre as variâncias amostrais: ", abs(variancia_amostral_estrategia_A - variancia_amostral_estrategia_B).round(3))
```

A partir dos valores da média e variância amostral, podemos retirar os seguintes *insights*:
1. Temos que as médias entre os alunos que utilizaram a estratégia A e alunos que utilizaram a estratégia B são muito próximas, ou seja, não há uma diferença significativa entre as médias amostrais. Esse fato confirma-se pelo fato da diferença absoluta ser relativamente pequena.
2. Em relação as variâncias, temos que em ambas as amostras, temos um valor relativamente alto, dessa forma, nota-se um grau de dispersão elevado para ambos. Além disso, temos que a diferença absoluta é relativamente alta, indicando que os estudantes que seguiram a estratégia A tiveram notas mais dispersas em relação aos estudantes que seguiram a estratégia B. A partir deste fato, pode-se indicar que estudantes que seguem a estratégia B têm notas menos discrepântes entre si do que os estudantes da estratégia A. 


# 3) Imprima os resultados da estatística do teste Z, p value e indique se rejeitaremos ou não a hipótese nula. Justifique sua resposta.

```python
n = len(amostra_estrategia_A)

normalizacao_z = (media_amostral_estrategia_B - media_amostral_estrategia_A) / (
    (variancia_amostral_estrategia_A/n + variancia_amostral_estrategia_B/n)**0.5
)

p_valor = 1 - stats.norm.cdf(normalizacao_z)

print("Estatística de teste Z:", normalizacao_z.round(3))
print("p-valor:", p_valor.round(3))

nivel_de_significancia = 0.05
if p_valor < nivel_de_significancia:
    print("Rejeitamos. Há evidências suficientes para concluir que os alunos que seguiram a estratégia B possuem uma nota média maior do que os alunos que seguiram com a estratégia A.")
else:
    print("Não rejeitamos. Não há evidências suficientes para concluir que os alunos que seguiram a estratégia B possuem uma nota média maior do que os alunos que seguiram com a estratégia A.")
```

Pré-determinamos o valor do nível de significância como $0.05$. Comparando este valor ao valor do p-value, $0.063$, temos que o valor do p-value é menor e, consequentemente, rejeitamos a hipótese nula.

# 4) Para finalizar monte o gráfico da distribuição da estatística do nosso teste Z e explique o que pode ser observado através dele.

```python
# Visualização da distribuição da estatística de teste Z e da região crítica
# Obs.: Segui o modelo da aula, adaptando para o caso da tarefa

fig, ax = plt.subplots(figsize=(10, 5))

x = np.linspace(-3, 3, 1000)
y = stats.norm.pdf(x, 0, 1)

ax.plot(x, y, label='Distribuição Normal Padrão')

ax.vlines(x=[normalizacao_z], ymin=0, ymax=stats.norm.pdf(normalizacao_z, 0, 1), colors='r', linestyles='dotted', label='Valor observado da estatística de teste')

critical_region = stats.norm.ppf(1 - nivel_de_significancia)

ax.fill_between(x, y, where=(x > critical_region), color='gray', alpha=0.3, label='Região crítica')

plt.legend()
plt.xlabel('Valor da Estatística de Teste Z')
plt.ylabel('Densidade')
plt.title('Distribuição da Estatística de Teste Z (teste unilateral à direita)')
plt.grid(True)
plt.show()
```
